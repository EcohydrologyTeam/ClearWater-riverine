```python
def setup_function_logger(name):
    logger = logging.getLogger('function_logger')
    handler = logging.FileHandler(f'{name}.log')
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    handler.setFormatter(formatter)
    logger.addHandler(handler)
    logger.setLevel(logging.DEBUG)  # Adjust the level based on your needs
    return logger
```

```python
def run_model(
    sim_name: str,
    boundary_condition_path: str | Path,
    initial_condition_path: str | Path,
    logging: bool = False,
    log_file_name: str = 'log',
    datetime_range = None,
):
    flow_field_fpath = f'W:/2ERDC12 - Clearwater/ClearwaterHECRAS_testCases/sumwereCreek_TSM_testing_timestep/clearWaterTestCases.{sim_name}.hdf'

    if logging:
        logger = setup_function_logger(f'{log_file_name}')

    if logging:
        logger.info(f'Starting {sim_name}.')

    # Instatiate the model object
    transport_model = cwr.ClearwaterRiverine(
        flow_field_fpath,
        diffusion_coefficient_input=0,
        verbose=True,
        datetime_range=datetime_range,
    )

    if logging:
        logger.info(f'Instantiated {sim_name}.')

    # Intialize the model
    transport_model.initialize(
        initial_condition_path=initial_condition_path,
        boundary_condition_path=boundary_condition_path,
        units='mg/m3',
    )

    if logging:
        logger.info(f'Initialized {sim_name}.')

    # simulate
    transport_model.simulate_wq(save=False,
        input_mass_units = 'g',
        input_volume_units = 'm3',
    )

    if logging:
        logger.info(f'Completed {sim_name} model run.')
    
    return transport_model
```

```python
def mb_run(transport_model, sim_name, save=False, save_name_add=''):
    mb_df_100_ans = _mass_bal_global_100_Ans(transport_model)
    mb_df_100_ans_t = mb_df_100_ans.transpose()
    mb_df = _mass_bal_global(transport_model)
    mb_df_t = mb_df.transpose()
    
    comparison_df = pd.DataFrame(
        {
            'Answer': mb_df_100_ans_t[0],
            'Simulated': mb_df_t[0]
        }
    )
    
    comparison_df['Diff'] = comparison_df['Answer'] - comparison_df['Simulated']
    comparison_df['CV'] = comparison_df['Diff'] / comparison_df['Answer']
    if save:
        comparison_df.to_csv(f'data/comparison_df_{sim_name}{save_name_add}.csv')
    return mb_df_t
    
```

### Why is Flow Coming into Coriolis Creek?

Start by examining the edges Jason identified as being associated with CC:

```python
face_flow = transport_model_30.mesh.face_flow.isel(
    nedge = 3899
)
face_flow
```

```python
face_flow.plot()
```

```python
face_flow = transport_model_30.mesh.face_flow.isel(
    nedge = 3800
)
face_flow.plot()
```

```python
face_flow = transport_model_30.mesh.face_flow.isel(
    nedge = 4093
)
face_flow.plot()
```

```python
face_flow = transport_model_30.mesh.face_flow.isel(
    nedge = 3802
)
face_flow.plot()
```

They all have negative face flow, which indicates that flow is coming in. No positive face flow (flow out). Let's look at the boundary conditions from the RAS HDF output:

```python
transport_model_30.boundary_data[transport_model_30.boundary_data['Name'] == 'CoriolisColdSpringCreek']
```

There are a lot more BCs associated with Coriolis Cold Spring Creek than expected! Let's take a look if any of them have positive face flow:

```python
edges = transport_model_30.boundary_data[transport_model_30.boundary_data['Name'] == 'CoriolisColdSpringCreek']['Face Index'].unique()
```

```python
import matplotlib.pyplot as plt
flow_out = []
for edge in edges:
    face_flow = transport_model_30.mesh.face_flow.isel(
    nedge = edge
    )
    face_flow.plot(label=edge)
    if (face_flow > 0).any():
        print(f"{edge}: positive!")
        flow_out.append(edge)
```

There are several edges that have postive flow! Let's take a look at which internal cells are associated with these:

```python
transport_model_30.mesh.edges_face1[flow_out]
```

Now plot where these fall on the map.

```python
fig, ax = plt.subplots()
sub_df = transport_model_30.gdf[transport_model_30.gdf.datetime == '2022-05-13 12:00:00']
sub_df.plot(ax=ax)
sub_df[sub_df.cell.isin([1653, 1660, 6771, 5907, 3998, 1968, 2276, 4306, 6215])].plot(ax=ax, edgecolor='red')
```

This is why we're seing flow goign out at CC: these cells are actually at the downstream boundary (where flow SHOULD be flowing out!)
